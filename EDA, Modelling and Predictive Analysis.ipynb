```r
df<- data13
```

```r
index = sample(1:nrow(df), 0.7*nrow(df)) 
train = df[index,] # Create the training data 
test = df[-index,] # Create the testing data
```

```r
install.packages("gradDescent", repos='http://cran.us.r-project.org')
```

```r
col_idx <- grep("Life.expectancy", names(train))
train2<- subset(train, select=c(2:9,1))
test2<- subset(test, select=c(2:9,1))
```

```r
library(gradDescent)
modelObject <- gradDescentR.learn(train2, featureScaling=TRUE, scalingMethod="VARIANCE",
                                  learningMethod="GD", control=list(alpha=0.05, maxIter=100), seed=NULL)
modelObject
```

```r
prediction_data <- predict(modelObject, train2)
prediction_test <- predict(modelObject, test2)
```

```r
eval_results <- function(actual, preds, df) {
  rss <- sum((preds - actual) ^ 2)
  tss <- sum((actual - mean(actual)) ^ 2)
  rsq <- 1 - rss/tss
  rsq
  rmse = sqrt(rss/nrow(df))
  data.frame(
    RMSE = rmse,
    Rsquare = rsq
  )
}
```

```r
eval_results(train2$Life.expectancy, prediction_data$V1, train2) # or use RMSE fn in the package
eval_results(test2$Life.expectancy, prediction_test$V1, test2)
```

```r
actuals_preds <- data.frame(cbind(actuals=test2$Life.expectancy, predicteds=prediction_test$V1))  # make actuals_predicteds dataframe.
cor(actuals_preds)  
head(actuals_preds)
```
